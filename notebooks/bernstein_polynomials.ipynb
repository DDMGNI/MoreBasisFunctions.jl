# Test Bernstein Polynomials

```julia
using MoreBasisFunctions
```

```julia
using Plots
```

```julia
plotlyjs()
```

Number of points to plot:

```julia
np = 101
```

Define Lagrange polynomial:

```julia
ber = Bernstein(5)
```

Compute plot range:

```julia
xp = collect(range(0,stop=1,length=np));
```

Plot all basis polynomials:

```julia
p = plot()
for i in 1:length(ber)
    yp = ber[i].(xp)
    plot!(p, xp, yp)
end
display(p)
```

```julia
p = plot()
yp = zero(xp)
for i in 1:length(ber)
    for j in eachindex(xp,yp)
        yp[j] = eval_element_derivative(ber, i, xp[j])
    end
    plot!(p, xp, yp)
end
display(p)
```
