# Test Lagrange Polynomials

```julia
using MoreBasisFunctions
```

```julia
using Plots
```

```julia
plotlyjs()
```

Create Lagrange basis:

```julia
lag = LagrangeGaussLegendre(5)
```

Number of points to plot:

```julia
np = 101
```

Compute plot range:

```julia
xp = collect(range(-1,stop=1,length=np));
```

Plot all basis polynomials:

```julia
p = plot()
for i in 1:length(lag)
    yp = lag[i].(xp)
    plot!(p, xp, yp)
end
display(p)
```

```julia
p = plot()
yp = zero(xp)
for i in 1:length(lag)
#    yp = lag'[i].(xp)
    for j in eachindex(xp,yp)
        yp[j] = eval_element_derivative(lag, i, xp[j])
    end
    plot!(p, xp, yp)
end
display(p)
```
